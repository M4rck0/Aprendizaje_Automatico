```python
# 1) DuraciÃ³n en minutos (para PelÃ­culas) y nÃºmero de temporadas (para Series)
minutos   = df["duration"].str.contains("min",    na=False)
temporada = df["duration"].str.contains("Season", na=False)

numero_extraido = pd.to_numeric(df["duration"].str.extract(r"(\d+)")[0], errors="coerce")

df["duracion_minutos"] = numero_extraido.where(minutos).astype("Float64")
df["numero_temporadas"] = numero_extraido.where(temporada).astype("Float64")
df
```

```python
df["pais_principal"] = df["country"].str.split(",").str[0].str.strip()
df
```

```python
df["es_pelicula"] = (df["type"] == "Movie")
df
```

```python
df = df.rename(columns={
    "show_id": "id_programa",
    "type": "tipo",
    "title": "titulo",
    "director": "director",
    "cast": "reparto",
    "country": "pais",
    "date_added": "fecha_agregado",
    "release_year": "anio_estreno",
    "rating": "clasificacion",
    "duration": "duracion",
    "listed_in": "generos",
    "description": "descripcion"
})
df
```

```python
# Eliminar columnas
df = df.drop(columns = ["es_pelicula", "descripcion"])
df
```

```python
# Filtrar datos
filtro_peliculas_largas = (
    (df["tipo"] == "Movie") &
    (df["duracion_minutos"] > 90) &
    (df["fecha_agregado"] >= "2019-01-01")
)
peliculas_largas = (
    df.loc[filtro_peliculas_largas,
           ["titulo","tipo","duracion_minutos","fecha_agregado","pais_principal","anio_estreno"]]
      .sort_values(["duracion_minutos","titulo"], ascending=[False, True])
      .reset_index(drop=True)
)
print("PelÃ­culas mayor a 90 minutos desde 2019:", len(peliculas_largas))
print(peliculas_largas.head(10))
print()
```

```python
# Crear grupos de datos

# Cantidad de tÃ­tulos por tipo y aÃ±o
grupo_tipo_decada = (
    df.groupby(["tipo","anio_estreno"], observed=True)
      .size()
      .rename("cantidad_titulos")
      .reset_index()
      .sort_values(["tipo","anio_estreno"])
)

print("Cantidad de tÃ­tulos por tipo y dÃ©cada:")
print(grupo_tipo_decada.head(10))
print()
```

```python
# Top 10 paÃ­ses con mÃ¡s tÃ­tulos
grupo_top_paises = (
    df.groupby("pais_principal", dropna=True, observed=True)
      .size()
      .rename("cantidad_titulos")
      .sort_values(ascending=False)
      .head(10)
)

print("Top 10 paÃ­ses con mÃ¡s tÃ­tulos en Netflix:")
print(grupo_top_paises)
print()
```
